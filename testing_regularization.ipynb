```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from ak_pynn.mlp import MLP
```

```python
X, Y = make_moons(100, noise=0.25,random_state=2)
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
model = MLP()
model.add_layer(2,input_layer=True)
model.add_layer(128,activation_function='relu')
model.add_layer(128,activation_function='relu')
model.add_layer(1,activation_function='sigmoid',output_layer=True)
model.compile_model(loss_function='binary_cross_entropy',optimizer='Adam',metrics=['accuracy'])
losses=model.fit(X, Y,epochs=500,verbose=False,learning_rate=0.01)
plot_decision_regions(X,Y,model,legend=2)
```

## L1_regularization

```python
from sklearn.metrics import accuracy_score
model = MLP()
model.add_layer(2,input_layer=True)
model.add_layer(128,activation_function='relu',L1_norm=0.0025)
model.add_layer(128,activation_function='relu',L1_norm=0.0025)
model.add_layer(1,activation_function='sigmoid',output_layer=True)
model.compile_model(loss_function='binary_cross_entropy',optimizer='Adam',metrics=['accuracy'])
losses=model.fit(X, Y,epochs=500,verbose=False,learning_rate=0.01,patience=3)
plot_decision_regions(X,Y,model,legend=2)
```

## L2_regularization

```python
from sklearn.metrics import accuracy_score
model = MLP()
model.add_layer(2,input_layer=True)
model.add_layer(128,activation_function='relu',L2_norm=0.03)
model.add_layer(128,activation_function='relu',L2_norm=0.03)
model.add_layer(1,activation_function='sigmoid',output_layer=True)
model.compile_model(loss_function='binary_cross_entropy',optimizer='Adam',metrics=['accuracy'])
losses=model.fit(X, Y,epochs=500,verbose=False,learning_rate=0.01,patience=3)
plot_decision_regions(X,Y,model,legend=2)
```

## Dropouts

```python
from sklearn.metrics import accuracy_score
model = MLP()
model.add_layer(2,input_layer=True)
model.add_layer(128,activation_function='relu',dropouts=True,drop_rate=0.65)
model.add_layer(128,activation_function='relu',dropouts=True,drop_rate=0.65)
model.add_layer(1,activation_function='sigmoid',output_layer=True)
model.compile_model(loss_function='binary_cross_entropy',optimizer='Adam',metrics=['accuracy'])
losses=model.fit(X, Y,epochs=500,verbose=False,learning_rate=0.01,patience=3)
plot_decision_regions(X,Y,model,legend=2)
```
